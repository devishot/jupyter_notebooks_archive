## Step by Step

![RAG](images/rag_create_and_index.png)
![RAG](images/rag_index_and_process.png)

```python
from langchain.document_loaders import CSVLoader
loader = CSVLoader(file_path=file)
```

```python
docs = loader.load()
```

```python
docs[0]
```

Note: Skip document chunking step as they already small

```python
from models import get_embeddings_model

embeddings = get_embeddings_model()
```

```python
embed = embeddings.embed_query("Hi my name is Harrison")
```

```python
print(len(embed))
print(embed[:5])
```

```python
db = DocArrayInMemorySearch.from_documents(
    docs, 
    embeddings
)
```

```python
query = "Please suggest a shirt with sunblocking"
```

```python
docs = db.similarity_search(query)
```

```python
print(len(docs))
print(docs[0])
```

```python
retriever = db.as_retriever()
```

```python
from models import get_chat_model

llm = get_chat_model(temperature=0.0)
```

```python
qdocs = "".join([docs[i].page_content for i in range(len(docs))])
```

```python
qdocs
```

```python
response = llm.invoke(f"{qdocs} Question: Please list all your \
shirts with sun protection in a table in markdown and summarize each one.").content
```

```python
display(Markdown(response))
```

## Same as above, but using Chains

![Stuff method](images/chain_stuff_method.png)
![Alternative methods](images/chain_map_reduce_and_refine_and_map_rerank.png)

```python
qa_stuff = RetrievalQA.from_chain_type(
    llm=llm, 
    chain_type="stuff", 
    retriever=retriever, 
    verbose=True
)
```

```python
query =  "Please list all your shirts with sun protection in a table \
in markdown and summarize each one."
```

```python
response = qa_stuff.run(query)
```

```python
display(Markdown(response))
```
